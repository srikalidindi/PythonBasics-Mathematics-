```
#Using definition
def fibonacci(n_terms):
  fib=[1,1]
  if n_terms<2:
    return fib[0:n_terms]

  for i in range(n_terms-2):
    element=fib[-2]+fib[-1]
    fib.append(element)
    return fib
```

```
#it is also possible to have a function without a return value
def show_fibonnaci(n_terms):
  for terme in fibonacci(n_terms):
    print(terme)
show_fibonnaci(12)
```

```
#defining default function as follows
def fibonacci_general(n_terms, first_term=1,second_term=1):
  fib=[first_term,second_term]
  if n_terms<2:
    return fin[0:n_terms]
  for i in range(n_terms-2):
    element=fib[-2]+fib[-1]
    fib.append(element)
  return fib
```

```
class vector:
  def __init__(self,size,value_initial=0):
    self.size=size
    self.elements=[value_initial]*size

  def show(self):
    print(self.elements)
  
  def copy(self):
    new=vector(self.size)
    new.elements=self.elements.copy()
    return new
  def access(self,index):
    return self.elements[index]
  def modif(self,index,value):
    self.elements[index]=value
  def addition(self,value):
    for i in range(self.value):
      self.elements[i]+=value
  
  def norm(self):
    cumul=0.0
    for x in self.elements:
      cumul += x**2
    return cumul**.5
```
