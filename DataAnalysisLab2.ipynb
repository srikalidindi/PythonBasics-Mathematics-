```
%matplotlib inline
import matplotlib.pyplot as plt
x = np.linspace(-5,5,100)
y=(6+4*x)/3
plt.plot(x,y,'-r',label='y')
plt.title('graph1')
plt.xlabel('x', color='#1C2833')
plt.ylabel('y', color='#1C2833')
plt.legend(loc='upper left')
plt.grid()
plt.show()
```
